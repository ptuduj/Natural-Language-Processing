```python
import requests
import json
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.patches import ArrowStyle
```

```python
def query(parameters):
    url = 'http://api.slowosiec.clarin-pl.eu/plwordnet-api/' + parameters
    r = requests.get(url)
    return json.loads(r.text)

def get_word_senses(w):
    return query('senses/search?lemma=' + str(w))['content']

def get_synset_from_sense(sense_id):
    return query('senses/' + str(sense_id) + '/synset')['id']

def get_sense_from_synset(synset_id):
    return query('synsets/' + str(synset_id) + '/senses')


def extract_senses(content):
    return [(i['id'], i['domain']['description']) for i in content]

    
def find_synonyms(word):
    senses = extract_senses(get_word_senses(word))
    synsets = [get_synset_from_sense(sense[0]) for sense in senses]
    synsets_senses = [get_sense_from_synset(s) for s in synsets]
    synonyms = [sense['lemma'] for s in synsets_senses for sense in s if sense['lemma']['word'] != word]
    return list({v['id']:v for v in synonyms}.values())
```

```python
def relation_to(synset_id):
    return query('synsets/' + str(synset_id) + '/relations/to')
    
def relation_from(synset_id):
    return query('synsets/' + str(synset_id) + '/relations/from')

def get_all_relation(synset_id):
    return query('synsets/' + str(synset_id) + '/relations')
```

```python
#zad 4
word = 'wypadek drogowy'
first_sense_id = extract_senses(get_word_senses(word))[0][0]
synset_id = get_synset_from_sense(first_sense_id)

sense_from_synset = get_sense_from_synset(synset_id)[0]

ids = [synset_id]
hypernyms_closure = [[(sense_from_synset['id'], sense_from_synset['lemma'])]]

for id_i in ids: 
    candidates = relation_from(id_i)
    
    for c in candidates:
        if c['relation']['name'] == 'hiponimia':
            s_id = c['synsetTo']['id']
            ids.append(s_id)
            senses = get_sense_from_synset(s_id)
            hypernyms_closure.append([(sense['id'], sense['lemma']) for sense in senses])      

hypernyms_closure
```

```python
edges = []
edge_labels = {}
for i in range(1, len(hypernyms_closure)):
    v1 = ''.join([ '`' + hypernym[1]['word'] + '`' for hypernym in hypernyms_closure[i-1]])
    v2 = ''.join([ '`' + hypernym[1]['word'] + '`' for hypernym in hypernyms_closure[i]])
    edges += [(v1, v2)]
    edge_labels[(v1,v2)] = ''

G = nx.DiGraph()
G.add_edges_from(edges)

plt.figure(figsize=(10,10))
pos = nx.kamada_kawai_layout(G)
nx.draw(G, pos, labels = {node:node for node in G.nodes()}, arrowsize=30, arrowstyle=ArrowStyle("->"))
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
```

```python
#zad 5
word = 'wypadek'

def get_direct_hyponyms(word, meaning_idx = 0):
    senses = extract_senses(get_word_senses(word))
    first_sense_id = senses[meaning_idx][0]
    synset_id = get_synset_from_sense(first_sense_id)

    candidates = relation_from(synset_id)
    hyponyms = []
    for c in candidates:
        if c['relation']['name'] == 'hiperonimia':
            senses = get_sense_from_synset(c['synsetTo']['id'])
            hyponyms.append([(sense['lemma'],sense['id']) for sense in senses])
    return [h for l_hyponyms in hyponyms for h in l_hyponyms] 



hyponyms = get_direct_hyponyms(word)
for h in hyponyms:
    print(h[0]['word'])
```

```python
#zad 6
def get_second_degree_hyponyms(word):
    direct_hyponyms = get_direct_hyponyms(word)
    second_degreen_hyponyms = []
    for d_h in direct_hyponyms:
        word_ = d_h[0]['word']
        second_degreen_hyponyms.extend(get_direct_hyponyms(word_))
    return second_degreen_hyponyms

second_degreen_hyponyms = get_second_degree_hyponyms(word)
for s_d_h in second_degreen_hyponyms:
    print(s_d_h[0]['word'])
```

```python
#zad 7
lexemes1 = {'szkoda':2, 'strata':1, 'uszczerbek':1,'szkoda majątkowa':1,'uszczerbek na zdrowiu':1,'krzywda':1,'niesprawiedliwość':1,'nieszczęście':2}
lexemes2 = {'wypadek': 1, 'wypadek komunikacyjny':1, 'kolizja':2, 'zderzenie':2, 'kolizja drogowa':1, 'bezkolizyjny':2, 'katastrofa budowlana':1, 'wypadek drogowy':1}

```

```python
def draw_graph(lexems):   
    senses = [(l, extract_senses(get_word_senses(l)), i) for l, i in lexems.items()]
    senses_f = []
  
    for s in senses:
        w = s[0]
        if s[1]:
            sense = s[1][lexems[w] - 1]
            senses_f.append((w, sense))
    synsets =  [ (w,get_synset_from_sense(s[0])) for w,s in senses_f] #get rigght meaning

    synsets_dict = {}
    for syn in synsets:
        synsets_dict[syn[1]] = [syn[0]] 

    relations = {}
    for id_, words in synsets_dict.items():
        relations_to = relation_to(id_)
        t = []
        for relation in relations_to:
            type_  = relation['relation']['name']
            if type_ == 'hiponimia':
                from_ = relation['synsetFrom']["id"]
                to_ = relation['synsetTo']["id"]
                t.append((from_, to_, type_))
        relations[id_]  = (words, t)

    edges = []
    edge_labels = {}
    for id_, (words,r) in relations.items():
        if r:
            v1 = ' '.join(['`'+word+'`' for word in words])
            for from_,to_,type_ in r:
                try:
                    words_from,_ = relations[from_]
                    v2 = ' '.join(['`'+word+'`' for word in words_from])
                    edges.append((v1,v2))
                    edge_labels[(v1,v2)] = ''
                except KeyError: 
                    pass
    
    G = nx.DiGraph()
    G.add_edges_from(edges)

    plt.figure(figsize=(10,10))
    pos = nx.kamada_kawai_layout(G)
    nx.draw(G,pos, labels={node:node for node in G.nodes()}, arrowsize=20, arrowstyle=ArrowStyle("->"))
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)    
```

```python
draw_graph(lexemes1)
```

```python
#zad8
from math import log
import queue


def lc_measure_with_dist(dist,max_dist):
    return -log((dist + 1) / (2.0 * max_dist))

def find_relation_distance(syn_id1,syn_id2,max_dist = 10):
    if syn_id1 == syn_id2:
        return 0
    pq = queue.PriorityQueue()
    pq.put((1,syn_id1))
    visited_syn_ids = set()
    while not pq.empty():
        distance, syn_id = pq.get()
        if syn_id not in visited_syn_ids:
            visited_syn_ids.add(syn_id)
            #get all realations
            relations =get_all_relation(syn_id)
            for relation in relations:
                try:
                    from_syn_id = relation['synsetFrom']['id'] 
                    if from_syn_id == syn_id2:
                        return distance 
                    if distance < max_dist:
                        pq.put((distance+1,from_syn_id))
                except KeyError:
                    pass
                
                try:
                    to_syn_id = relation['synsetTo']['id']
                    if to_syn_id == syn_id2:
                        return distance 
                    if distance < max_dist:
                        pq.put((distance+1,to_syn_id))
                except KeyError:
                    pass
            
    return -1

def LC(pair,max_distance = 20):
    (word1,id1),(word2,id2) = pair
    syn1_id,syn2_id = (None,None)
    try: 
        syn1_id = get_synset_from_sense(extract_senses(get_word_senses(word1))[id1-1][0])
        syn2_id = get_synset_from_sense(extract_senses(get_word_senses(word2))[id2-1][0])
    except IndexError:
        raise ValueError('specified sense from word was not found')
    dist = find_relation_distance(syn1_id,syn2_id)
    return (lc_measure_with_dist(dist,max_distance),dist) if dist >= 0 else (-1,dist)

for pair in [(('szkoda',2), ('wypadek',1)) ,(('kolizja',2), ('szkoda majątkowa',1))\
              ,(('nieszczęście',2), ('katastrofa budowlana',1))]:
    try:
        lc,dist = LC(pair)
        print(' LC({},{}) = {} with distance {}'.format(pair[0],pair[1], lc,dist))
    except ValueError as e:
        print(e)
```
